```python
import requests
import datetime
import json
```

```python
POST_CODE = "713104"
age = 52

# Print details flag
print_flag = 'Y'

numdays = 20
```

```python
base = datetime.datetime.today()
date_list = [base + datetime.timedelta(days=x) for x in range(numdays)]
date_str = [x.strftime("%d-%m-%Y") for x in date_list]
```

```python
for INP_DATE in date_str:
    URL = "https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/calendarByPin?pincode={}&date={}".format(POST_CODE, INP_DATE)
    response = requests.get(URL)
    if response.ok:
        resp_json = response.json()
        flag = False
        json_data = json.loads(response.text)
        if resp_json["centers"]:
            for i in json_data["centers"]:
                for j in i["sessions"]:
                    if j["min_age_limit"] <= age:
                        flag = True
                        break
                if flag:
                    break
            if not flag:
                continue
            print("Available on: {}".format(INP_DATE))
            if print_flag=='y' or print_flag=='Y':
                print("++Center details:")
                for i in json_data["centers"]:
                    print("\tName:", i["name"], "[ID:", i["center_id"], "]")
                    print("\t++Sessions:")
                    for j in i["sessions"]:
                        print("\t\tVaccine:", j["vaccine"])
                        print("\t\tCapacity:", j["available_capacity"])
                        print("\t\tSlots available:", j["slots"])
            print("======================================================================\n")
```
